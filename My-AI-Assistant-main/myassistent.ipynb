```python
def ask_question(name):
  ques = 'hey'+ name + 'what do you want'
  ques = input("ques")
  return ques
```

```python
def classify_questions(ques):
  gohead_with_web_search = False
  device_list =['alarm','remainder','message','call']
  personal_list = {'who are you ': 'Iam jarvis','who created you':'iam created by reddi'}
  device = False
  for i in device_list:
    if i in ques:
      device = True

    if device:
      print(" this question is related to device things")

    personal_question = False
    for i in personal_list:
       if i in ques.lower():
           personal_question = True

    if personal_question:
        print("hey iam jarvis")


    if device:
      gohead_with_web_search = False
    elif personal_question:
      gohead_with_web_search = False
    else:
      gohead_with_web_search = True

    return gohead_with_web_search




```

```python
classify_questions("Who are you ")
```

```python
import pathlib
import textwrap

import google.generativeai as genai

from IPython.display import display
from IPython.display import Markdown
from IPython.display import Audio, display



def to_markdown(text):
  text = text.replace('â€¢', '  *')
  return Markdown(textwrap.indent(text, '> ', predicate=lambda _: True))

```

```python
# Or use `os.getenv('GOOGLE_API_KEY')` to fetch an environment variable.
GOOGLE_API_KEY=userdata.get('GOOGLE_API_KEY')

genai.configure(api_key=GOOGLE_API_KEY)
```

```python
model = genai.GenerativeModel('gemini-pro')
```

```python
response = model.generate_content("What is the meaning of life?")
```

```python
def ask_gemini(ques):
  modified_prompt = 'hey give me answer to this question' + ques + 'in maximum of 40 words'
  response = model.generate_content(ques)
  to_markdown(response.text)
  return response.text
```

```python
def speak(answer):
  tts = gTTS(answer)
  tts.save('audio.mp3')
  display(Audio('audio.mp3', autoplay=True))
```

```python

have_any_other_ques = 'y'
name = ''

while have_any_other_ques == 'y':
  if name == '':
    name = input("hey what is your name")
    q = ask_question(name)

  go_ahead = classify_questions(q)
  answer = ''

  if go_ahead == True:
    answer = ask_gemini(q)
    speak(answer)
    print (answer)







  have_any_other_ques = input( " do yoy need me ?")
```
